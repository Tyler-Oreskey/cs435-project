```python
from landsatxplore.earthexplorer import EarthExplorer
from landsatxplore.api import API
from dotenv import load_dotenv
import pandas as pd
import os
from pathlib import Path
```

```python
load_dotenv()

username = os.environ["LANDSATXPLORE_USERNAME"].strip()
password = os.environ["LANDSATXPLORE_PASSWORD"].strip()

ee = EarthExplorer(username, password)
```

```python
lat, lng = 78.754, -41.871
landsat_9 = ee.api.search(
    dataset="landsat_ot_c2_l2",
    latitude=lat,
    longitude=lng,
    max_cloud_cover=5,
    months=[9]
)

landsat_8 = ee.api.search(
    dataset="landsat_ot_c2_l1",
    latitude=lat,
    longitude=lng,
    max_cloud_cover=5,
    months=[9],
)

df = pd.DataFrame(landsat_8 + landsat_9)
df
```

```python
years = df.groupby(df["acquisition_date"].dt.year).first()#.reset_index(drop=True)

for id, dataset in years.iterrows():
    if Path("./data/" + dataset["display_id"] + ".tar").exists():
        print("Already all downloaded", dataset["display_id"])
        continue
    ee.download(dataset["entity_id"], output_dir="./data")
    print("Downloaded", dataset["display_id"])
```
